```python
#Creat dataframe
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'data': {1: 13 , 2: 12, 3: 32},
    'Weight': {1: 10, 2: 20, 3: 30}},
    columns=['data', 'Weight']
)
df.head()
```

```python
#find the mean for the data columne
df.data.mean()
```

```python
#weighted mean  is measure of center where some values are counted more than once
weightedmean= (df.data * df.Weight).sum() / df.Weight.sum()
weightedmean
```
