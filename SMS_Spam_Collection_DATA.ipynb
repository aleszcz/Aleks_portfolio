```python
#string punctuation all the characters are printed out
import string
string.punctuation
```

```python
#removing punctuation using list comprehantion
def remove_punct_char(txt):
  txt_nopunct = [c for c in txt if c not in string.punctuation]
  return txt_nopunct
```

```python
# Assuming remove_punct_char is a defined function to remove punctuation characters
#problem is convrting to character we need words
#next code is to help with this problem
data['msg_clean'] = data['msg'].apply(lambda x: remove_punct_char(x))
data.head()
```

```python
# IMPORTANT modification remove punctuation character and combine words

def remove_punctuation(txt):
  txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
  return txt_nopunct

data['msg_clean'] = data['msg'].apply(lambda x: remove_punctuation(x))
data.head()

```

```python
# 3 Tokenizaton process now converting clean text to tokens
# using REg expression with \W not a Word Character  using this simple tokenizer
import re
def tokenize (txt):
  tokens = re.split('\W+', txt)
  return tokens

data['msg_clean_tokenized'] = data['msg_clean'].apply(lambda x: tokenize(x.lower()))
data.head()
```

```python
#4 removing stop words
#call in NLTK library
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
#text_tokenized is the parameter being feed to this function
def remove_stopwords(text_tokenized):
    txt_clean = [word for word in text_tokenized if word not in stopwords]
    return txt_clean

# calling in the Lambda fucntion
# look into the the generated text
data['msg_no_sw'] = data['msg_clean_tokenized'].apply(lambda x: remove_stopwords(x))
data.head()
```

```python
# 5 Stemming and lemmatization
# using porter STemmer

import nltk
from nltk.stem import PorterStemmer
ps= PorterStemmer()
#dir(ps)


# print examples
print(ps.stem('coder'))
print(ps.stem('coding'))
print(ps.stem('code'))
```

```python
#Effect of stemming on the text

def stemming (tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text


# list comprehansion and lambda function
data[ ' msg_stemmed'] = data['msg_no_sw'].apply(lambda x: stemming(x))
data.head()
```

```python
# Lemmatization inflected forms of a word to be analyzed as a single root word or lemma
import nltk
wn = nltk.WordNetLemmatizer()
nltk.download('wordnet')
#dir(wn)
```

```python
#example
print('lemmatizing')
print(wn.lemmatize('cactus'))
print(wn.lemmatize('cacti'))
```

```python
from ast import Lambda
#defninig the lemmatization function
def lemmatization(token_txt):
    text = [wn.lemmatize(word) for word in token_txt]
    return text


data['msg_lemmatized'] = data['msg_no_sw']. apply (lambda x: lemmatization(x))
data.head()
```

```python
# 5 Vectorization using different methods: Count +nGram vectorizer, or vectorization using TF-IDF, Word-Vec
# Scikit-learn in Count Vectorizer

import string
import re
import nltk

nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
#text Cleaning
def clean_text(txt):
    txt = "".join([c for c in txt if c not in  string.punctuation])
    tokens = re.split('\W+',txt )
    txt = [ps.stem(word) for word in tokens if word not in stopwords]
    return txt
```

```python
# Count vectorizer

from sklearn.feature_extraction.text import CountVectorizer
cv1 = CountVectorizer(analyzer=clean_text)
X = cv1.fit_transform(data['msg'])
print(X.shape)
```

```python
# nGram Vectorizer
# Scikit-Learn Count Vectorizer that uses nGrams
import string
import re
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
def clean_text_ngram(txt):
    txt = "".join([c for c in txt if c not in string.punctuation])
    tokens = re.split('\W+', txt)
    ps = PorterStemmer()
    txt = " ".join([ps.stem(word) for word in tokens if word not in stopwords.words('english')])
    return txt
```
