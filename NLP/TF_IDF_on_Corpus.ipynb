```python
# Term Frequency on Corpus
#NLTK tokenizer and collection/Counter
#Computed Counts+TF
from collections import Counter
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
```

```python
import pandas as pd
import numpy as np

import string
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
#Count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
# read coprus
corpus = ['the man went for a walk', 'the children sat around the fire']
```

```python
X = cv.fit(corpus)

print(cv.get_feature_names_out())
print(X.vocabulary_)
```

```python

Y = cv.transform(corpus)
Y = cv.fit_transform(corpus)
print(Y.shape)
print(Y)
print(Y.toarray())
```

```python
# converting count to Df and display
df = pd.DataFrame(Y.toarray(), columns = cv.get_feature_names_out())
print(df)
```

```python
df.T
```

```python
#count the frequences
sumDocu = np.zeros(Y.shape[0])
docFreq = np.zeros(Y.shape)

for i in range(len(corpus)):
  sumDocu[i] = sum(Y.toarray()[i])
  print (sumDocu[i])
  for j in range(len(Y.toarray()[i])):
    docFreq[i][j] = Y.toarray()[i][j]/sumDocu[i]

print(docFreq)
```

```python
dfFreq = pd.DataFrame(docFreq, columns = cv.get_feature_names_out())
print(dfFreq)
```
