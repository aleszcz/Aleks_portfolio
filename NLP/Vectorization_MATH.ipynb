```python
import numpy as np

# Define the normalized vectors
A_norm = np.array([3/np.sqrt(58), 7/np.sqrt(58)])
B_norm = np.array([5/np.sqrt(1177), 36/np.sqrt(1177)])
C_norm = np.array([7/np.sqrt(850), 29/np.sqrt(850)])
D_norm = np.array([212/np.sqrt(44928), 72/np.sqrt(44928)])
E_norm = np.array([23/np.sqrt(533), 2/np.sqrt(533)])

# Create a list of normalized vectors
normalized_vectors = [A_norm, B_norm, C_norm, D_norm, E_norm]

# Compute Cosine similarity and Cosine distance between all pairs of normalized vectors
cosine_similarities = np.zeros((len(normalized_vectors), len(normalized_vectors)))
cosine_distances = np.zeros((len(normalized_vectors), len(normalized_vectors)))

for i in range(len(normalized_vectors)):
    for j in range(len(normalized_vectors)):
        dot_product = np.dot(normalized_vectors[i], normalized_vectors[j])
        norm_product = np.linalg.norm(normalized_vectors[i]) * np.linalg.norm(normalized_vectors[j])

        cosine_similarity = dot_product / norm_product
        cosine_distance = 1 - cosine_similarity

        cosine_similarities[i, j] = cosine_similarity
        cosine_distances[i, j] = cosine_distance

# Print the Cosine similarities and Cosine distances
print("Cosine Similarities:")
for i in range(len(normalized_vectors)):
    for j in range(len(normalized_vectors)):
        print(f"{labels[i]} to {labels[j]}: {cosine_similarities[i, j]}")

print("\nCosine Distances:")
for i in range(len(normalized_vectors)):
    for j in range(len(normalized_vectors)):
        print(f"{labels[i]} to {labels[j]}: {cosine_distances[i, j]}")
```

```python
import nltk
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import gutenberg
from collections import Counter, OrderedDict

nltk.download('gutenberg')

def get_clean_word_list(text):
    # Tokenize and filter punctuations
    puncs = set((',', '.', '--', '-', '!', '?', ':', ';', '``', "''", '(', ')', '[', ']'))
    word_list = (x.lower() for x in nltk.word_tokenize(text) if x not in puncs and '_' not in x and ' ' not in x)
    return list(word_list)

def plot_zipfs_distribution(word_list, title):
    # Count word frequency
    token_counts = Counter(word_list)
    # Sort token counts
    sorted_token_counts = OrderedDict(sorted(token_counts.items(), key=lambda kv: kv[1], reverse=True))

    freqVector = np.array(list(sorted_token_counts.values()))
    rank = np.arange(1, len(sorted_token_counts) + 1)

    # Plot Zipf's distribution
    plt.figure(figsize=(10, 6))
    plt.plot(np.log10(rank), np.log10(freqVector), marker='o', linestyle='', label=title)
    plt.title('Zipf\'s Distribution - ' + title)
    plt.xlabel('log(Rank of Words)')
    plt.ylabel('log(Frequency of Words)')
    plt.legend()
    plt.show()

```

```python
# Download and process Hamlet and Romeo & Juliet
hamlet_text = gutenberg.raw('/content/Hamlet.txt')
romeo_juliet_text = gutenberg.raw('/content/RomeoJuliet.txt')

hamlet_word_list = get_clean_word_list(hamlet_text)
romeo_juliet_word_list = get_clean_word_list(romeo_juliet_text)

# Plot Zipf's distribution for Hamlet
plot_zipfs_distribution(hamlet_word_list, 'Hamlet')

# Plot Zipf's distribution for Romeo & Juliet
plot_zipfs_distribution(romeo_juliet_word_list, 'Romeo & Juliet')
```

```python

#using NLTK tokenizer

import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from sklearn.feature_extraction.text import CountVectorizer

# Download NLTK resources
nltk.download('punkt')

# Set up CountVectorizer
vect = CountVectorizer()

# Read corpus
corpus = ["disneyland is in orange county", "orange county disneyland is better than orlando disneyland"]

# Tokenize and create FreqDist
tokenized_corpus = [word_tokenize(sentence) for sentence in corpus]
flat_list = [word.lower() for sublist in tokenized_corpus for word in sublist]
freq_dist = FreqDist(flat_list)

# Calculate TF for each document
tf_values = []
for sentence in tokenized_corpus:
    tf_doc = {word: freq / len(sentence) for word, freq in freq_dist.items() if word in sentence}
    tf_values.append(tf_doc)

# Print TF values
print(tf_values)
```

```python
#using count vectorizer
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

# Set up CountVectorizer
vect = CountVectorizer()

# Read corpus
corpus = ["disneyland is in orange county", "orange county disneyland is better than orlando disneyland"]

# Fit and transform using CountVectorizer
X = vect.fit_transform(corpus)

# Get feature names and vocabulary
feature_names = vect.get_feature_names_out()
vocabulary = vect.vocabulary_

# Calculate TF values
tf_values = []

for i in range(X.shape[0]):
    doc_tf = {}
    for j, term_index in enumerate(X[i].indices):
        term = feature_names[term_index]
        doc_tf[term] = X[i, term_index] / X[i].sum()

    tf_values.append(doc_tf)

# Print TF values
print(tf_values)
```
