```python
#measure the lenght of the sentences
#compute the total number of characters
#count all the sentances that have 280 characters
numChars = 280
count= 0
for s in blobTotal.sentences:
  if len(s) > numChars:
    #print(s, len(s))
    count += 1
    ##########################33

print('Total number of snetences that have more than 280 charcacters =', count)
```

```python
# give the entences that contain certian words : like Romeo and Juliet
stringCh1= 'Romeo'
stringCh2 = 'Juliet'
count = 0
for s in blobTotal.sentences:
  if (stringCh1 in s) and (stringCh2 in s):
    print(s)
    print()
    count +=1
print(count)
```

```python
#Display longest and shortes sentences
from datascience import *
list_Length = []
for s in blobTotal.sentences:
  list_Length.append(len(s))

t1 = Table().with_columns('Sentence', blobTotal.sentences, 'Length', list_Length)
##########################################################
#Long sentences
sortedTableLong = t1.sort('Length', descending=True)
sortedTableLong.show(10)
############################################################
#shortest sentences
sortedTableShort = t1.sort('Length', descending=False)
sortedTableShort.show(10)
```
