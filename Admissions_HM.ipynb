```python
# 1. Load the libraries
import pandas as pd
from sklearn import preprocessing

# 2. Read the Dataset
data = pd.read_csv('Admissions.csv')

# Select the features you want to scale
features = data[['gre', 'gpa', 'rank']]

# Scale the selected features between 0 and 1
featuresScale = preprocessing.minmax_scale(features)

# Display the first 5 rows of the scaled features
print(featuresScale[0:5])
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Load Libraries
# Done in the previous steps

# 2. Read Dataset
data = pd.read_csv('/content/Admissions.csv')

# 3. Scale the predictor variable values between 0 and 1
scaler = MinMaxScaler()
data[['gre', 'gpa']] = scaler.fit_transform(data[['gre', 'gpa']])

# 4. Encode the response categorical variable into one-hot variable
data['Admit_One-Hot-1'] = 1 - data['admit']
data['Admit_One-Hot-2'] = data['admit']
data = data.drop(columns=['admit'])

# 5. Split data into “Training” and “Testing”
X = data[['gre', 'gpa', 'rank']]
y = data[['Admit_One-Hot-1', 'Admit_One-Hot-2']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6. Build the Neural Network Model Using Keras/TF software
model = Sequential()
model.add(Dense(5, input_dim=3, activation='relu'))
model.add(Dense(2, activation='softmax'))

# 7. Train the Model using “Training” data
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=32)

# 8. Print values of all the weights and bias of all the neurons of the Neural Network
for layer in model.layers:
    weights, bias = layer.get_weights()
    print(f"Layer Weights: {weights}")
    print(f"Layer Bias: {bias}")

# 9. Predict the outcome using the “Testing” data
y_pred = model.predict(X_test)

# 10. Build confusion matrix
y_pred_class = np.argmax(y_pred, axis=1)
y_test_class = np.argmax(np.array(y_test), axis=1)
confusion = confusion_matrix(y_test_class, y_pred_class)

# 11. Compute the accuracy of the prediction
accuracy = accuracy_score(y_test_class, y_pred_class)

# Print the confusion matrix and accuracy
print(f"Confusion Matrix:\n{confusion}")
print(f"Accuracy: {accuracy}")
```
